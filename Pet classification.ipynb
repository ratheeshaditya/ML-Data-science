<h2>Importing the required library's</h2>

```
import seaborn as sns
import cv2
from random import shuffle
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```
TRAIN_PATH = "train/"
TEST_PATH = "test/"
IMG_SIZE = (100,100)
```

*   Iterating through each class and assigning label's accordingly 1 for dog , 0 for cat
*   Reading the image and resizing to the img size
*   Appending the label and training data to the corresponding variables




```
def process_img():
  training_data =[]
  label_data = []
  for classes in os.listdir(TRAIN_PATH):
    path = os.path.join(TRAIN_PATH,classes)
    for img in os.listdir(path):
      label = 1 if classes=="dogs" else 0
      new_img = cv2.imread(os.path.join(path,img))
      new_img = cv2.resize(new_img , IMG_SIZE)
      training_data.append(np.array(new_img))
      label_data.append(label)
  return np.array(training_data),label_data
```

```
X , Y = process_img()
```

```
X.shape
```

Converting the label's to categorical format so it can be fed to the convolutional neural net

```
Y = to_categorical(Y,2)
```

<h2>Model building</h2>

<h3>Importing the required library's</h3>

```
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from keras.optimizers import Adam
from tensorflow.keras.layers import Softmax
```

*   Training the model with the learning rate of 1e-5
*   And Optimizer : Adam



```
INPUT_SHAPE = (100,100,3)
learning_rate = 1e-5
optimizer = Adam(lr=learning_rate)
```

```
model = Sequential()

model.add(Conv2D(32, (5, 5), input_shape=(INPUT_SHAPE),padding='same'))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64,(5,5),padding='same'))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2,2)))

model.add(Flatten())
model.add(Dense(32))
model.add(Dropout(0.4))
model.add(Dense(2))
model.add(Activation("softmax"))
```

```
model.compile(loss="categorical_crossentropy",
              optimizer=optimizer,
              metrics=["accuracy"])
```

<h3>Summary of the model</h3>

```
model.summary()
```

For evaluation purposes we will import confusion matrix as well as accuracy_score from sklearn

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

<h3>Processing test data</h3>

```
def process_test_data():
  test_data = []
  true_labels = []
  for classes in os.listdir(TEST_PATH):
    path = os.path.join(TEST_PATH,classes)
    for files in os.listdir(path):
      labels = 1 if classes=="dogs" else 0
      new_img = cv2.imread(os.path.join(path,files))
      new_img = cv2.resize(new_img,(IMG_SIZE))
      test_data.append(np.array(new_img))
      true_labels.append(labels)
  return np.array(test_data),true_labels
```

```
TEST_INPUT_SHAPE = (-1,100,100,3)
```

```
test , true_label = process_test_data()
```

```
test_data = test.reshape(TEST_INPUT_SHAPE)
```

<h2>Training the model</h2>

<h3>For 100 Epochs</h3>

```
model.fit(X,Y,epochs=100,steps_per_epoch=X.shape[0])
```

```
prediction = [np.argmax(model.predict(img.reshape(TEST_INPUT_SHAPE))) for img in test_data]
```

```
print(confusion_matrix(true_label,prediction))
```

```
print (f"The accuracy is : {accuracy_score(true_label,prediction)*100}%")
```
